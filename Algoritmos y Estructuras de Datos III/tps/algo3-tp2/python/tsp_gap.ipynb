## GAP iteraciones vs mejores vecinos

```python
# Calculamos gap de TS para Berlin52 para distintos numeros de iteraciones y de mejores vecinos explorados.

it = [i*2 for i in range(1, 50)] # Nro de iteraciones
it_size = len(it)

num_vecinos = [1, 3, 5]

sol_optima = 7542

gap_ts = []

for i in num_vecinos:
    for j in it:
        res = ejecutar("berlin52", 'TS', 50, i, j) # [tiempo, peso]
        gap_ts.append(float(res[1]-sol_optima)/sol_optima)

ts_1 = gap_ts[0: it_size]
ts_3 = gap_ts[it_size: it_size*2]
ts_5 = gap_ts[it_size*2: it_size*3]
#ts_7 = gap_ts[it_size*3: it_size*4]
#ts_9 = gap_ts[it_size*4: it_size*5]
```

```python
def time_gap_ts():
    
    fig, ax = plt.subplots()
    ax.plot(it, ts_1, label="TS - 1 vecino")
    ax.plot(it, ts_3, label="TS - 3 vecinos")
    ax.plot(it, ts_5, label="TS - 5 vecinos")
    #ax.plot(it, ts_7, label="TS - 7 vecinos")
    #ax.plot(it, ts_9, label="TS - 9 vecinos")
    
    ax.legend()
    ax.set(xlabel='iteraciones', ylabel='gap',
           title='TS | iteraciones vs gap | memoria = 50')
    ax.grid()

    fig.savefig("graficos/gap_vecinos_ts")
    
time_gap_ts()
```

## GAP iteraciones vs memoria

```python
# Calculamos gap de TS para Berlin52 para distintos numeros de iteraciones y tamao de memoria.

it = [i*2 for i in range(1, 50)] # Nro de iteraciones
it_size = len(it)

tam_memoria = [20, 50, 100]

sol_optima = 7542

gap_ts = []

for i in tam_memoria:
    for j in it:
        res = ejecutar("berlin52", 'TS', i, 1, j) # [tiempo, peso]
        gap_ts.append(float(res[1]-sol_optima)/sol_optima)

ts_20 = gap_ts[0: it_size]
ts_50 = gap_ts[it_size: it_size*2]
ts_100 = gap_ts[it_size*2: it_size*3]
```

```python
def mem_gap_ts():
    
    fig, ax = plt.subplots()
    ax.plot(it, ts_20, label="TS - M = 20")
    ax.plot(it, ts_50, label="TS - M = 50")
    ax.plot(it, ts_100, label="TS - M = 100")
    
    ax.legend()
    ax.set(xlabel='iteraciones', ylabel='gap',
           title='TS | iteraciones vs gap | vecinos = 1')
    ax.grid()
    
    fig.savefig("graficos/gap_memoria_ts")

mem_gap_ts()
```
